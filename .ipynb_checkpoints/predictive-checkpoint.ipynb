```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

def preprocess_data(dataset):
    # Check for missing values
    missing_values = dataset.isnull().sum()
    print(f"Missing values before handling: \n{missing_values}")

    # Separate numeric and categorical columns
    numeric_cols = dataset.select_dtypes(include=[np.number]).columns
    categorical_cols = dataset.select_dtypes(include=[object]).columns

    # Handle missing values for numeric columns (fill with median)
    dataset[numeric_cols] = dataset[numeric_cols].fillna(dataset[numeric_cols].median())
    
    # Handle missing values for categorical columns (fill with mode)
    for col in categorical_cols:
        dataset[col] = dataset[col].fillna(dataset[col].mode()[0])  # Fixed here

    print(f"Missing values after handling: \n{dataset.isnull().sum()}")

    # Feature Engineering: Create rolling average for sensor data
    if 'sensor_1' in dataset.columns:  # Check if 'sensor_1' exists
        dataset['sensor_rolling_avg'] = dataset['sensor_1'].rolling(window=5).mean()

    # Normalize only numeric columns
    if len(numeric_cols) > 0:
        scaler = MinMaxScaler()
        dataset[numeric_cols] = scaler.fit_transform(dataset[numeric_cols])

    # Optionally, handle categorical columns (e.g., encoding)
    if len(categorical_cols) > 0:
        print(f"Categorical columns detected: {categorical_cols}")
        # For example, you could apply Label Encoding or One-Hot Encoding
        for col in categorical_cols:
            if dataset[col].nunique() <= 2:  # If the column has only two unique values (binary categorical), use Label Encoding
                dataset[col] = dataset[col].map({'No Failure': 0, 'Failure': 1})  # Example: 'No Failure' -> 0, 'Failure' -> 1
            else:  # Otherwise, apply One-Hot Encoding
                dataset = pd.get_dummies(dataset, columns=[col])

    return dataset

# Example Usage
dataset = pd.read_csv(r"C:\Users\BineshMahato\.cache\kagglehub\datasets\shivamb\machine-predictive-maintenance-classification\versions\1\predictive_maintenance.csv")
processed_dataset = preprocess_data(dataset)

# Check the updated dataset
processed_dataset.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming `dataset` is the preprocessed dataset and you have removed non-numeric columns already
# Select only numeric columns for the correlation matrix
numeric_cols = dataset.select_dtypes(include=[np.number]).columns

# Calculate the correlation matrix for numeric columns
corr_matrix = dataset[numeric_cols].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar_kws={'label': 'Correlation Coefficient'})

# Show the plot
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Assuming 'dataset' is the preprocessed dataset

# Step 1: One-hot encode categorical columns
# We will drop 'Target' column from X (features) and use it for y (target)
X = dataset.drop(columns=['Target'])  # Drop the target column
y = dataset['Target']  # Use 'Target' as the target variable

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Step 2: Check for any non-numeric columns (if any)
print(X_encoded.dtypes)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Step 4: Train Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 5: Make predictions
y_pred = rf_model.predict(X_test)

# Step 6: Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Check for NaN or infinite values
print(X_encoded.isna().sum())  # Check for NaN values
print((X_encoded == float('inf')).sum())  # Check for infinite values
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Failure', 'Failure'], yticklabels=['No Failure', 'Failure'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
